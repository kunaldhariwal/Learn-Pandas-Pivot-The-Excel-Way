```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)

# disable ssl temporarily if certificates file is missing and you get an ssl error
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
#load the dataset
movies = pd.read_csv('https://raw.githubusercontent.com/kunaldhariwal/Learn-Pandas-Pivot-The-Excel-Way/main/pivot_excel_way_dataset.csv',index_col=0)
```

```python
movies.head()
```

```python
# basic Pivot 1
pd.pivot_table(movies, index='Year',values='Movie',aggfunc='count')
```

```python
# basic Pivot 2
pd.pivot_table(movies, index='LeadStudio',columns='Year',values='Movie',aggfunc='count')
```

```python
# fill NaN in the results above

pd.pivot_table(movies, index='LeadStudio',columns='Year',values='Movie',aggfunc='count', fill_value=0)
```

```python
# basic Pivot 3
pd.pivot_table(movies, index='Genre',values='Profitability',aggfunc='sum', fill_value=0)
```

```python
# Advance Pivot 1
pd.pivot_table(movies, index='LeadStudio',values=['Movie','AudienceScore'], aggfunc={'Movie':'count', 'AudienceScore':'mean'}, fill_value=0)
```

```python
# Advance Pivot 2
pd.pivot_table(movies, index='LeadStudio',values=['DomesticGross','ForeignGross','WorldGross','Budget','Genre'], aggfunc={'DomesticGross':'max', 'WorldGross':'min','Budget':['min','max','mean'],'Genre':"nunique"}, fill_value=0)
```

```python
# Advance Pivot 3
pd.pivot_table(movies[movies.LeadStudio.isin(['Sony','DreamWorks','Disney'])], index='LeadStudio',values=['DomesticGross','ForeignGross','WorldGross','Budget','Genre'], aggfunc={'DomesticGross':'max', 'WorldGross':'min','Budget':['min','max','mean'],'Genre':"nunique"}, fill_value=0)
```
